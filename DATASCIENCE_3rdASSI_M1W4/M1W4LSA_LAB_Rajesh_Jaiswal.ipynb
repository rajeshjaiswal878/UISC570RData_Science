#### In this step will fetch data from Baseball group in to 'rajcorpus'

```python
from sklearn.datasets import fetch_20newsgroups
categories = ['rec.sport.baseball']
dataset = fetch_20newsgroups(subset='all',shuffle=True, random_state=42, categories=categories)
rajcorpus = dataset.data #Data fetched from baseball group news into local data set 'rajcorpus'
```

```python
#Data of rajcorpus
rajcorpus
```

#### In next few steps will collect reducndent word of html page and remove them

```python
from bs4 import BeautifulSoup
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
```

```python
#Download stopward dictonary
nltk.download('stopwords')
```

#### In this step will update stopword file

```python
stopset = set(stopwords.words('english'))
stopset.update(['lt','p','/p','br','amp','quot','field','font','normal','span','0px','rgb','style','51', 
                'spacing','text','helvetica','size','family', 'space', 'arial', 'height', 'indent', 'letter'
                'line','none','sans','serif','transform','line','variant','weight','times', 'new','strong', 'video', 'title'
                'white','word','letter', 'roman','0pt','16','color','12','14','21', 'neue', 'apple', 'class',  ])
```

### In this step we will work toward TF-IDF vectorizing

We are going to use scikit-learn's TF-IDF vectorizer to convert each document into a sparse matrix of TFIDF.

```python
#Data of 1st document in corpus
rajcorpus[0]
```

### In this step will create sparse matrix

```python
rajvectorizer = TfidfVectorizer(stop_words=stopset,
                                 use_idf=True, ngram_range=(1, 3))
rajX = rajvectorizer.fit_transform(rajcorpus)
```

```python
#1st Documents matrix presentation will tell us number word it have
rajX[0]
```

```python
#print sparse matrix 'rajX'
print (rajX[0])
```

```python
#Details of full rajX matrix with number of documents and total words count of rajcorpus.
rajX.shape
```

### In this step will apply LSA technique

Input: rajX, a matrix where m is the number of documents, and n is the number of word.
         Matrix rajX  m=994,n=188088 

Process: Applied  decomposition  on rajX and divided into three matricies called rajU, rajS, and rajV. 
When we do the decomposition, we have to pick a value k, that's how many concepts we are going to keep.

We are going to consider first 36 Concepts there for k=36.
<li>Hence,rajU[m,k]=rajU[994,36]
      <li>rajS[k,k]=rajS[36,36]
      <li>rajV[n,k]=rajV[188088,36] transpose matrix

```python
#Considering concepts n=36
rajlsa = TruncatedSVD(n_components=36, n_iter=100)
rajlsa.fit(rajX)
```

```python
#1st row of component
rajlsa.components_[0]
```

```python
#Will Check version of python and gcc 
import sys
print (sys.version)
```

## ------------------------------------------------------------------- 
### Final Step :To generate concepts report of first 36 rows

```python
rajterms = rajvectorizer.get_feature_names()
for i, comp in enumerate(rajlsa.components_): 
    termsInComp = zip (rajterms,comp)
    sortedTerms =  sorted(termsInComp, key=lambda x: x[1], reverse=True) [:10]
    print ("Concept %d:" % i)
    for term in sortedTerms:
        print (term[0])
    print (" ")
```
